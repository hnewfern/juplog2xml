## Convert Log to XML

```python
startTag = 'FAILED' #@param{type: 'string'}
endTag = 'END FAILURE' #@param{type: 'string'}
startnode1 = '<failure>' #@param{type: 'string'}
endnode1 = '</failure>' #@param{type: 'string'}

import re

timestamp=re.compile(r'[0-9]{4}-[0-9]{2}-[0-9]{2}T\.{1}[0-9]{2}:[0-9]{2}:[0-9]{2}.[0-9]{7}Z$')
tsLine=re.compile(r'[0-9]{4}-[0-9]{2}-[0-9]{2}T\.{1}[0-9]{2}:[0-9]{2}:[0-9]{2}.[0-9]{7}Z\s')
tscount=0
nodecount=0
innode=False;
logFile=open('data/logFile.txt', 'r')
logLine=logFile.readline()

print("<?xml version=\"1.0\" encoding=\"UTF-8\"?>")
print("<log>")
while logLine:
    logLine = logLine.rstrip()
    if re.match(timestamp, logLine):
        if tscount<1:
            logLine=logFile.readline()
            tscount+=1
        elif tscount==1:
            logLine=logFile.readline()
            tscount=0
    else:
        if re.findall(startTag, logLine):
            print('   ' + startnode1)
            print('      ' + logLine)
            nodecount+=1
            innode=True
            logLine=logFile.readline()
        elif re.findall(endTag, logLine):
            print('      ' + logLine)
            print('   ' + endnode1)
            innode=False
            logLine=logFile.readline()
        elif nodecount>0 and innode==True:
            result=re.sub(tsLine, r'', logLine)
            print('      ' + result)
            logLine=logFile.readline()
            tscount=0
        else:
            logLine=logFile.readline()
            tscount=0
print("</log>")
            
        
```
