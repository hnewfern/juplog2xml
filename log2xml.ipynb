
## Convert Log to XML

```python
startTag = 'FAILED' #@param{type: 'string'}
endTag = 'END FAILURE' #@param{type: 'string'}
startnode1 = '<failure>' #@param{type: 'string'}
endnode1 = '</failure>' #@param{type: 'string'}
startnode1child1='<title>'
endnode1child1='</title>'
startnode1child2='<description>'
endnode1child2='</description>'

import ipywidgets as widgets
import re as regex

#timestamp=regex.compile(r'[0-9]{4}-[0-9]{2}-[0-9]{2}T\.{1}[0-9]{2}:[0-9]{2}:[0-9]{2}.[0-9]{7}Z$')
tsLine=regex.compile(r'[0-9]{4}-[0-9]{2}-[0-9]{2}T\.{1}[0-9]{2}:[0-9]{2}:[0-9]{2}.[0-9]{7}Z')
#tscount=0
nodecount=0
innode=False;


logFile=open('data/logFile.txt', 'r')
logLine=logFile.readline()

print("<?xml version=\"1.0\" encoding=\"UTF-8\"?>")
print("<log>")
while logLine:
    logLine = logLine.rstrip()
    result=regex.sub(r'[\s]+', ' ', logLine)
    if regex.findall(startTag, result):
        
        print('   ' + startnode1 + '\n      ' + startnode1child1 + result + endnode1child1)
        print('      ' + startnode1child2)
        nodecount+=1
        innode=True
        logLine=logFile.readline()
    elif nodecount>0 and innode==True:
        if regex.findall(endTag, result):
            result=regex.sub(tsLine, '', endTag)
            print('      ' + result + '\r      ' + endnode1child2)
            print('\r   ' + endnode1)
            innode=False
            logLine=logFile.readline()
        else: 
            result=regex.sub(tsLine, '', logLine)
            print('      ' + result)
            logLine=logFile.readline()            
#    elif nodecount>0 and innode==False:
#            logLine=logFile.readline()
        #tscount=0
    else:
        logLine=logFile.readline()
        #tscount=0
print("</log>")
            
        
```
